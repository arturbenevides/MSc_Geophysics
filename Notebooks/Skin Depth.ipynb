## Referencial Teórico

obs: Aqui partimos objetivamente de onde surge o conceito de skin depth, para uma maior explanação ver texto do projeto de pesquisa.

Considerando o módulo do campo elétrico

$$
|E_{x}(Z)|=|E_{x}^{s}|e^{-\sqrt{\frac{\omega \mu \sigma}{2}}Z}
$$

Isso decai monotonicamente com o aumento de z. A profundidade na qual $E_{x}|$ decresce a partir da superfície (z = 0) por um fator $1/e$ é definida pelo **skin depth** (**$\delta$**)

$$
\frac{|E_{x}(z=\delta)|}{E_{x}^{s}} = \frac{1}{e} = e^{-\frac{\omega\mu\sigma}{2}\delta}
$$

que equivale:
    
$$
\sqrt{\frac{\omega\mu\sigma}{2}}\delta= 1
$$

que pode ser rearranjado para dar:
    
$$
\delta=\sqrt{\frac{2}{\omega\mu\sigma}}
$$    


Substituindo $\mu$ por $\mu_{0}=4\pi 10^{-7}$ da uma expessão para o skin depth em metros, tal como:

$$
\delta=\frac{503}{\sqrt{\sigma f}}
$$

ou 
$$
\delta=503\sqrt{\frac{\rho}{f}}
$$

## Simulações

Abaixo simulamos 3 situações distintas em que temos frequências iguais =  1 Hz e 100 Hz, para um meio com resistividade uniforme igual a 10 ohmm e outro com 1000 ohm.m

```python
z=np.linspace(100000, 0, 1001) # Considerando um range de orofundidade escolhidas 0 até a profundidade de 8km com cada metro amostrado.
#v=np.linspace(1000, 0, 1001)
mi=4*np.pi*10E-07 

sig1=1/10 #condutividade
```

### Frequencia= 0.1 Hz

```python
freq103=1000
delta13= np.exp(-np.sqrt(freq103*mi*sig1/2)*z)
```

### Frequêcia = 1 Hz

```python
freq=1
delta1= np.exp(-np.sqrt(freq*mi*sig1/2)*z)
```

### Frequêcia = 10 Hz

```python
freq10=10
delta10= np.exp(-np.sqrt(freq10*mi*sig1/2)*z)
```

### Frequêcia = 100 Hz

```python
frequencia2=100
delta2= np.exp(-np.sqrt(frequencia2*mi*sig1/2)*z)
```

### Frequêcia = 1000 Hz

```python
freq01=0.1
delta01= np.exp(-np.sqrt(freq01*mi*sig1/2)*z)
```

# Meio homogêneo com 10 ohm.m

```python
fig = plt.figure(figsize=(7.0, 7.0))


freq1000 = plt.plot(delta13, z, label="$10^{3} Hz$", 
                 linestyle='solid', color='blue', linewidth=1)

freq1 = plt.plot(delta2, z, label="$10^{2} Hz$", 
                   linestyle='solid', color='yellow', linewidth=1)
freq10 = plt.plot(delta10, z, label="$10^{1} Hz$", 
                   linestyle='solid', color='red', linewidth=1)

freq100 = plt.plot(delta1, z, label="$10^{0} Hz$", 
                   linestyle='solid', color='black', linewidth=1)

#freq01 = plt.plot(delta01, v, label="0.1 Hz",linestyle='solid', color='green', linewidth=1)

plt.legend()

plt.ylabel('Profundidade $(m)$', fontsize=14)
plt.xlabel('$E_{x} / E_{x}^{0}$', fontsize=14)

plt.text(0.85,7800,"$10^{1} ohm.m$", fontsize=10)
plt.yticks(np.arange(0, 100001, 1000))
#plt.yscale('log')
#plt.yticks(Profundidade (km), ["0.5", "1.0", "1.5", "2.0", "2.5", "3.0"])  
           
pylab.ylim([10000,0])
plt.show()

fig.savefig('skindepth10.png', dpi=600)
```
